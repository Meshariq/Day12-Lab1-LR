## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_boston,y_boston, test_size=0.4, random_state=100)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

## Q9: Train the Model.

```python
lm.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lm.intercept_)
```

```python
x = data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
coeff_df = pd.DataFrame(lm.coef_, x.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
Real_Values = np.array(y_test)
Real_Values
```

```python
plt.scatter(Real_Values,predictions);
```

## Residual Histogram

```python
sns.distplot((y_test-predictions),bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
